```r
library(ArchR)
library(rtracklayer)
addArchRThreads(threads = 16)
```

```r
proj <- loadArchRProject(path = "/mnt/g/genomics/10_16_23_CA1/Allsamples/")
proj
```

```r
markerGenes <- c("Ndrg1","Dcn","Dlk1","Adra1a","Ntng1",
                 "Lct","Olfm3","Wfs1","Epha7","Klhl13")

p <- plotEmbedding(
    ArchRProj = proj, 
    colorBy = "GeneScoreMatrix", 
    name = markerGenes, 
    embedding = "UMAP",
    imputeWeights = getImputeWeights(proj),
    quantCut = c(0.01,0.99)
)

```

```r
#Rearrange for grid plotting
p2 <- lapply(p, function(x){
    x + guides(color = FALSE, fill = FALSE) + 
    theme_ArchR(baseSize = 5) +
    theme(plot.margin = unit(c(0, 0, 0, 0), "cm")) +
    theme(
        axis.text.x=element_blank(), 
        axis.ticks.x=element_blank(), 
        axis.text.y=element_blank(), 
        axis.ticks.y=element_blank()
    )
})

p3 <- do.call(cowplot::plot_grid, c(list(ncol = 5),p2))
```

```r
p3
```

```r
ggsave(filename = "DV_genes.png",p3,width = 8,height = 4,dpi=600)
```
