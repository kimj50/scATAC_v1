```python
# read gtf file as bed9 and returns bed6

def gtf_to_bed6(gtf):
    gtf_1 = gtf[gtf['end'] == 'gene']
    c1 = gtf_1['chrom']
    c2 = gtf_1['name']
    c3 = gtf_1['score']
    c4 = gtf_1['rgb'].str.split(';',expand=True)[0].str.replace('gene_id ', '').str.replace('"', '')
    c5 = gtf_1['strand']
    c6 = gtf_1['thickStart']
    c7 = gtf_1['rgb'].str.split(';',expand=True)[2].str.replace('gene_name ', '').str.replace('"', '').astype(str).str.replace(' ', '')
    BED = pd.concat([c1,c2,c3,c4,c5,c6,c7],axis=1)
    BED['strand'] = 1000
    BED.columns = ['chrom','start','end','name','score','strand','gene_name']
    BED_sorted = bioframe.sort_bedframe(BED)
    return BED_sorted

```
